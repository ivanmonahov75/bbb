```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
import math
```

```python
name = "2"
plt.rcParams.update({'font.size': 22})
```

```python
data = pd.read_csv(name + ".csv", names=["C", "nu"])
```

```python
X = data["C"].values
sigma_X = 0.2
Y = 1 / data["nu"].values * 1000
sigma_Y = 0.2
```

```python
A = np.vstack([X, np.ones(len(X))]).T
k, b = np.linalg.lstsq(A, Y, rcond=None)[0]
#sigma_k = math.sqrt((Y.std() - Y.mean()**2 ) / (X.std() - X.mean()**2 ) - b**2 ) / math.sqrt(len(X))
#eps_k = sigma_k / k
#sigma_b = sigma_k * math.sqrt(X.std() / X.mean() - X.mean()**2 )
```

```python
fig = plt.figure(figsize=(12, 7))
ax = fig.gca()
plt.scatter(X, Y, marker=".")
plt.errorbar(X, Y, xerr=sigma_X, yerr=sigma_Y, linestyle="None")
delta_x = (X.max() - X.min()) / len(X)
delta_y = (Y.max() - Y.min()) / len(Y)
ax.set_xlim(0, 60)
ax.set_ylim(0, 5)
plt.xlabel("$C, 10^{-3} мкФ$")
plt.ylabel("$T, 10^{-3} c$")
plt.plot(X, (k*X + b), 'r', label='Fitted line')
plt.plot(X, (0.07881 * X - 0.015))
plt.grid(True)
plt.savefig("../" + name + ".png")
```
