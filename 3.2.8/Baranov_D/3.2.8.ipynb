```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
i = np.array([
2.94,
3.93,
5.30,
7.14,
9.23,
11.1,
13.5,
16.4,
14.6,
12.4,
9.90,
6.90,
4.56,
2.47,
1.7])

u = np.array([
87.9 ,
92.6 ,
100.0,
110.2,
120.0,
130.2,
143.0,
158.0,
148.5,
136.7,
123.5,
108.7,
95.3 ,
84.5 ,
82.4])
```

```python
plot_1 = plt.figure(figsize=(12*0.9,10*0.9))
plt.grid(visible=True, linewidth=0.6)

plt.title('Зависимость $I(U)$', fontsize=18)
#plt.xlim(xmin=0, xmax=0.03)
#plt.ylim(ymin=0, ymax=100)
plt.ylabel('$I$, мА', fontsize=20, rotation=0, ha='right')
plt.xlabel('$U$, В', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=16, size=10)
plt.grid(visible=True, linewidth=0.6)



yerr_1 = 0.05
xerr_1 = 0.5



#z, err = np.polyfit(D, Q[1], deg=1, cov=True)
#polynom_2 = np.poly1d(z)
#plt.plot(D, polynom_2(D),"b--", linewidth=1, label="МНК")

plt.errorbar(
    u[:7],
    i[:7],
    fmt='rs',
    linewidth=1,
    markersize=1,
    elinewidth=1,
    label='$I(U)$, увеличение U',
    xerr=xerr_1,
    #yerr=yerr_1,
    capsize=3
)

plt.errorbar(
    u[8:],
    i[8:],
    fmt='bs',
    linewidth=1,
    markersize=1,
    elinewidth=1,
    label='$I(U)$, уменьшение U',
    xerr=xerr_1,
    #yerr=yerr_1,
    capsize=3
)


plt.errorbar(
    u[:] - i[:]*5.1,
    i[:],
    fmt='ks',
    linewidth=1,
    markersize=1,
    elinewidth=1,
    label='$I(U - rI)$',
    xerr=xerr_1,
    #yerr=yerr_1,
    capsize=3
)



#z = np.polyfit(1/p, D, deg=1)
#polynom_1 = np.poly1d(z)
#p = np.array([40, 100, 160, 220, 280, 760])
#plt.plot(1/p, polynom_1(1/p),"r--", linewidth=1, label="МНК")


plt.legend(fontsize=18, markerscale = 1)
plt.show()

plot_1.savefig(
    'I(U).pdf',
    format='pdf',
    bbox_inches='tight',
    #pad_inches=4
)
```
